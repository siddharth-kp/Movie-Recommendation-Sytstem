```python
import nltk
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
```

```python
def filter_stop_words(text):
    stop_words = set(stopwords.words('english'))
    words = nltk.word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    filtered_text = ' '.join(filtered_words)
    return filtered_text
```

```python
word_set={"set_not_dict"}
for i in range(len(dataset.iloc[:,0])):
#     genres
    for j in json.loads(dataset.iloc[i][0]):
        word=j['name']
        dataset.iloc[i][len(dataset.iloc[0,:])-1].append(word)
        word_set.add(word)
#     keywords
    for j in json.loads(dataset.iloc[i][2]):
        word=j['name']
        dataset.iloc[i][len(dataset.iloc[0,:])-1].append(word)
        word_set.add(word)
#      overview
    filtered_sent= filter_stop_words(dataset.iloc[i][3])
    for j in filtered_sent:
        dataset.iloc[i][len(dataset.iloc[0,:])-1].append(word)
        word_set.add(word)
#      spoken_languages
    for j in json.loads(dataset.iloc[i][5]):
        word=j['name']
        dataset.iloc[i][len(dataset.iloc[0,:])-1].append(word)
        word_set.add(word)
#       Tagline
    filtered_sent= filter_stop_words(dataset.iloc[i][6])
    for j in filtered_sent:
        dataset.iloc[i][len(dataset.iloc[0,:])-1].append(word)
        word_set.add(word)
#     Title
    dataset.iloc[i][len(dataset.iloc[0,:])-1].append(dataset.iloc[i][7])
    word_set.add(dataset.iloc[i][7])
    
dataset.head(1)
```
